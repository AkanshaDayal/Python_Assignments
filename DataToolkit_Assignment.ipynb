#Theory Questions:

1. What is NumPy, and why is it widely used in Python?
  - NumPy is a powerful library for numerical computing in Python, offering fast array operations and support for multidimensional arrays.

2. How does broadcasting work in NumPy?
  - Broadcasting allows NumPy to perform operations on arrays of different shapes by automatically expanding their dimensions to match.

3. What is a Pandas DataFrame?
  - A DataFrame is a 2D labeled data structure in Pandas, similar to a table or spreadsheet.

4. Explain the use of the groupby() method in Pandas?
  - groupby() splits data into groups based on some criteria, then applies a function like sum or mean on each group.

5. Why is Seaborn preferred for statistical visualizations?
  - Seaborn simplifies complex visualizations and comes with built-in themes and functions for statistical plots.

6. What are the differences between NumPy arrays and Python lists?
  - NumPy arrays are faster, more memory-efficient, and support vectorized operations, unlike Python lists.

7. What is a heatmap, and when should it be used?
  - A heatmap visualizes data using color gradients, ideal for showing correlations or matrix-like data.

8. What does the term “vectorized operation” mean in NumPy?
  - It refers to applying operations over entire arrays without explicit loops, leading to faster execution.

9. How does Matplotlib differ from Plotly?
  - Matplotlib is static and script-based, while Plotly supports interactive, web-based plots.

10. What is the significance of hierarchical indexing in Pandas?
  - It enables multiple levels of indexing, allowing complex data structures like multi-level rows and columns.

11. What is the role of Seaborn’s pairplot() function?
  - pairplot() creates scatter plots between all numerical variable pairs, useful for exploring relationships.

12. What is the purpose of the describe() function in Pandas?
  - It provides summary statistics (mean, count, std, etc.) for numerical columns in a DataFrame.

13. Why is handling missing data important in Pandas?
  - Missing data can skew analysis results; handling it ensures accuracy and consistency.

14. What are the benefits of using Plotly for data visualization?
  - Plotly offers interactive, zoomable, and web-embeddable visualizations with minimal code.

15. How does NumPy handle multidimensional arrays?
  - NumPy uses the ndarray object, allowing efficient storage and operations on arrays of any dimension.

16. What is the role of Bokeh in data visualization?
  - Bokeh is used for creating interactive and real-time plots in web applications.

17. Explain the difference between apply() and map() in Pandas?
  - map() works element-wise on Series, while apply() can apply functions to Series or DataFrame rows/columns.

18. What are some advanced features of NumPy?
  - Advanced features include broadcasting, masked arrays, memory-mapped files, and linear algebra routines.

19. How does Pandas simplify time series analysis?
  - It offers built-in support for date indexing, resampling, rolling windows, and frequency conversion.

20. What is the role of a pivot table in Pandas?
  - Pivot tables summarize and reorganize data, making it easier to analyze grouped data.

21. Why is NumPy’s array slicing faster than Python’s list slicing?
  - NumPy slices return views instead of copies and are implemented in optimized C code.

22. What are some common use cases for Seaborn?
  - Common uses include visualizing distributions, correlations, categorical data, and regression models.

#Practical Questions:

```python
#1. How do you create a 2D NumPy array and calculate the sum of each row?
import numpy as np
arr = np.array([[1, 2, 3], [4, 5, 6]])
row_sum = np.sum(arr, axis=1)
print(row_sum)
```

```python
#2.  Write a Pandas script to find the mean of a specific column in a DataFrame.
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
mean_b = df['B'].mean()
print(mean_b)
```

```python
#3.  Create a scatter plot using Matplotlib.
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 15, 13, 17]
plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')
plt.show()
```

```python
#4.  How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import seaborn as sns
import pandas as pd
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [5, 6, 7],
    'C': [8, 9, 10]
})
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
#5.  Generate a bar plot using Plotly.
import plotly.express as px
data = {'Fruits': ['Apple', 'Banana', 'Orange'], 'Quantity': [10, 15, 7]}
df = pd.DataFrame(data)
fig = px.bar(df, x='Fruits', y='Quantity', title='Fruit Quantity')
fig.show()
```

```python
#6. Create a DataFrame and add a new column based on an existing column.
df = pd.DataFrame({'A': [1, 2, 3]})
df['B'] = df['A'] * 2
print(df)
```

```python
#7.  Write a program to perform element-wise multiplication of two NumPy arrays.
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
result = a * b
print(result)
```

```python
#8.  Create a line plot with multiple lines using Matplotlib.
x = [1, 2, 3]
y1 = [2, 4, 6]
y2 = [1, 3, 5]
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.legend()
plt.title('Multiple Lines')
plt.show()
```

```python
#9.  Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold. df = pd.DataFrame({'A': [10, 20, 5, 30]})
filtered = df[df['A'] > 15]
print(filtered)
```

```python
#10.  Create a histogram using Seaborn to visualize a distribution.
import seaborn as sns
import numpy as np
data = np.random.randn(100)
sns.histplot(data, kde=True)
plt.title('Distribution')
plt.show()
```

```python
#11.  Perform matrix multiplication using NumPy.
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
product = np.dot(a, b)
print(product)
```

```python
#13.  Create a 3D scatter plot using Plotly.
import plotly.express as px
import pandas as pd
df = pd.DataFrame({
    'x': [1, 2, 3],
    'y': [4, 5, 6],
    'z': [7, 8, 9]
})
fig = px.scatter_3d(df, x='x', y='y', z='z', title='3D Scatter Plot')
fig.show()
```
